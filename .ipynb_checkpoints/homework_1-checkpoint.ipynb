```python
import av
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset, DataLoader
import time
import torch.multiprocessing as mp
```

```python
try:
    mp.set_start_method('spawn', force=True)
except RuntimeError:
    pass
```

```python
video_file = "video.mp4"
```

# Задание 1

```python
def read_clip(filename, start=0, num_frames=16, stride=2):
    with av.open(filename) as container:
        stream = container.streams.video[0]
        stream.thread_type = 'AUTO'
        
        frames = []
        sampled_indices = []
        
        current_idx = 0
        collected_count = 0
        
        for frame in container.decode(stream):
            if current_idx >= start:
                if (current_idx - start) % stride == 0:
                    img = frame.to_ndarray(format='rgb24')
                    frames.append(img)
                    sampled_indices.append(current_idx)
                    collected_count += 1
            
            if collected_count == num_frames:
                break
                
            current_idx += 1
                
        fps = stream.average_rate
    
    return np.array(frames), sampled_indices, float(fps)
```

```python
start_frame = 5
n_frames = 40
stride_step = 10

clip, indices, fps = read_clip(video_file, start=start_frame, num_frames=n_frames, stride=stride_step)

print(f"Shape: {clip.shape}")
print(f"FPS: {fps}")
print(f"Indices: {indices}")

if len(clip) > 0:
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    axes[0].imshow(clip[0])
    axes[0].set_title(f"First Frame (Index {indices[0]})")
    axes[0].axis('off')

    axes[1].imshow(clip[-1])
    axes[1].set_title(f"Last Frame (Index {indices[-1]})")
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()
```

# Задание 2

```python
class VideoDataset(Dataset):
    def __init__(self, video_paths, clip_len=16, stride=2, transform=None):
        self.video_paths = video_paths
        self.clip_len = clip_len
        self.stride = stride
        self.transform = transform

    def __len__(self):
        return len(self.video_paths)

    def __getitem__(self, idx):
        path = self.video_paths[idx]
        frames, _, _ = read_clip(path, start=0, num_frames=self.clip_len, stride=self.stride)
        
        if self.transform:
            frames = self.transform(frames)
        
        return frames

def to_tensor_transform(frames):
    # frames shape: (T, H, W, 3) - numpy.ndarray, uint8
    # Convert to float tensor and normalize
    frames = torch.from_numpy(frames).float() / 255.0
    # Permute to (T, C, H, W) format, suitable for DataLoader stacking (B, T, C, H, W)
    frames = frames.permute(0, 3, 1, 2)
    return frames
```

```python
video_files = [f"video_dataset/video{i+1}.mp4" for i in range(2)]

dataset = VideoDataset(
    video_files, 
    clip_len=16, 
    stride=2
)

loader = DataLoader(dataset, batch_size=4, shuffle=False)

print(f"Dataset length: {len(dataset)}")

for batch in loader:
    print(f"Batch shape: {batch.shape}")
    print(f"Batch type: {type(batch)}")
    break
```

# Задание 3

```python
def run_benchmark(dataset, num_workers_list, batch_size=4):
    throughputs = []
    
    print(f"{'Workers':<10} | {'Time (s)':<10} | {'Throughput (fps)':<20}")
    print("-" * 45)

    for nw in num_workers_list:
        loader = DataLoader(dataset, batch_size=batch_size, num_workers=nw, shuffle=False)
        
        start_time = time.time()
        total_frames = 0
        
        for batch in loader:
            total_frames += batch.shape[0] * batch.shape[1]
            
        end_time = time.time()
        duration = end_time - start_time
        fps = total_frames / duration
        throughputs.append(fps)
        
        print(f"{nw:<10} | {duration:<10.4f} | {fps:<20.2f}")
        
    return throughputs
```

```python
n_files = 2

video_files = [f"video_dataset/video{i+1}.mp4" for i in range(2)]

dataset = VideoDataset(
    video_files, 
    clip_len=16, 
    stride=2,
    transform=to_tensor_transform
)

workers_list = [0, 1, 2, 4, 8]
results = run_benchmark(dataset, workers_list, batch_size=4)

plt.figure(figsize=(10, 6))
plt.plot(workers_list, results, marker='o', linewidth=2)
plt.title("DataLoader Throughput vs Num Workers")
plt.xlabel("Number of Workers")
plt.ylabel("Throughput (Frames / sec)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(workers_list)

for i, txt in enumerate(results):
    plt.annotate(f"{txt:.1f}", (workers_list[i], results[i]), 
                 xytext=(0, 10), textcoords='offset points', ha='center')
    
plt.show()
```
